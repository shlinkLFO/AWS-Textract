```python
import cv2
import numpy as np

def draw_gridlines(input_path, output_path):
    # Load the image
    image = cv2.imread(input_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply adaptive thresholding
    thresh = cv2.adaptiveThreshold(
        gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2
    )

    # Detect horizontal lines
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (40, 1))
    horizontal_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)

    # Detect vertical lines
    vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 20))
    vertical_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, vertical_kernel, iterations=2)

    # Combine horizontal and vertical lines to get the grid
    grid = cv2.add(horizontal_lines, vertical_lines)

    # Convert grid to 3-channel image (to overlay on the original image)
    grid_colored = cv2.cvtColor(grid, cv2.COLOR_GRAY2BGR)

    # Overlay the grid on the original image
    combined = cv2.addWeighted(image, 0.8, grid_colored, 0.2, 0)

    # Save the output image
    cv2.imwrite(output_path, combined)

# Example usage
input_image_path = 'C:\\Users\\mason\\td\\td\\td\\grid_lines.png'
output_image_path = 'C:\\Users\\mason\\td\\td\\td\\grid_lines_with_output.png'
draw_gridlines(input_image_path, output_image_path)
```

```python
import cv2
import numpy as np

def draw_fixed_gridlines(input_path, output_path, horizontal_lines, vertical_lines):
    # Load the image
    image = cv2.imread(input_path)

    # Create a copy to draw gridlines
    grid_image = image.copy()

    # Draw horizontal lines
    for y in horizontal_lines:
        cv2.line(grid_image, (0, y), (grid_image.shape[1], y), (0, 0, 255), 1)  # Red lines for visibility

    # Draw vertical lines
    for x in vertical_lines:
        cv2.line(grid_image, (x, 0), (x, grid_image.shape[0]), (0, 0, 255), 1)  # Red lines for visibility

    # Save the output image
    cv2.imwrite(output_path, grid_image)

# Example usage
input_image_path = 'C:\\Users\\mason\\td\\td\\td\\grid_lines.png'
output_image_path = 'C:\\Users\\mason\\td\\td\\td\\grid_lines_with_fixed_output.png'

# Specify exact pixel locations for horizontal and vertical gridlines
horizontal_lines = [35, 65, 95, 120, 150, 180, 205, 231, 258, 290, 317, 345, 370]  # Example y-coordinates
vertical_lines = [50, 95, 155, 205, 255, 305, 355, 410, 445, 490, 550, 600, 650, 700, 750]  # Example x-coordinates

draw_fixed_gridlines(input_image_path, output_image_path, horizontal_lines, vertical_lines)
```
